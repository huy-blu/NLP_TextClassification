```python
import pandas as pd
from selenium import webdriver
driver= webdriver.Chrome('../chromedriver.exe')
driver.get('https://vnexpress.net/suc-khoe/dinh-duong')
datalink={}
link={}
stt=1
def crawlLink(stt):
    contain= driver.find_elements_by_css_selector('h2.title-news')
    for p in contain:
        link=driver.find_element_by_link_text(p.text).get_attribute('href')
        datalink[stt]=link
        stt+=1
    if(stt>1000):
        driver.close()
    else:
        step = driver.find_element_by_xpath('//a[@class="btn-page next-page "]')
        step.click()
        crawlLink(stt)
crawlLink(stt)
```

```python
# import pandas as pd
df = pd.DataFrame(data=datalink, index=[0])
df = (df.T)
df.columns=['Link']
df.to_excel('../data/datalinkSucKhoe.xlsx')
```
